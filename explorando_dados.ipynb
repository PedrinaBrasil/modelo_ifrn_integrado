```python
#import os pacotes utilizados
import numpy as np
import pandas as pd

#exibindo todas as colunas
pd.set_option('display.max_columns', None)
```

```python
dados = pd.read_json('dados/dados_flat.json')
```

```python
dados
```

```python
dados.shape
```

```python
dados.groupby(['sigla', 'descricao_historico_disciplina'])['media_final'].aggregate([np.mean]).unstack()
```

```python
df = dados.loc[dados['descricao_historico_disciplina'].isin(['Língua Portuguesa e Literatura I(90H)',
                                                        'Língua Portuguesa e Literatura II(90H)', 
                                                        'Matemática I(120H)', 'Matemática II(90H)'])]

df.groupby(['sigla', 'descricao_historico_disciplina'])['media_final'].aggregate([np.mean]).unstack()
```

```python
dados = df
```

```python
dados.groupby(['descricao_historico_disciplina', 'descricao'])['media_final'].aggregate([np.mean, np.size]).unstack()
```

```python
dados.groupby(['descricao_historico_disciplina', 'descricao', 'sigla'])['media_final'].aggregate([np.size]).unstack()
```

```python
df = dados.loc[dados['descricao'].isin(['Evasão',
                                        'Cancelado', 
                                        'Matriculado'])]

df.groupby(['descricao_historico_disciplina', 'descricao', 'sigla'])['media_final'].aggregate([np.size]).unstack()
```

```python
#df[~df['id'].duplicated()].groupby(['descricao', 'sigla', 'ano_letivo__ano'])['id'].value_counts()
df[~df['id'].duplicated()].groupby(['descricao', 'sigla', 'ano_letivo__ano'])['id'].aggregate([np.size]).unstack()
```

```python
#agrupando os alunos cancelados, evadidos, transferidos interno e externo como evasão para o mec
#-1 valor sentinela, 0 evasão, 1 matriculado
dados['classe'] = np.ones(dados.shape[0])*-1
```

```python
dados.loc[dados['descricao'].isin(['Evasão',
                                   'Cancelado', 
                                   'Cancelamento Compulsório', 
                                   'Transferido Externo', 
                                   'Transferido Interno']), 'classe'] = 0

dados.loc[dados['descricao'].isin(['Concluído',
                                   'Intercâmbio', 
                                   'Matriculado', 
                                   'Trancado', 
                                   'Trancado Voluntariamente', 
                                   'Matrícula Vínculo Institucional']), 'classe'] = 1
```

```python
dados.groupby('classe').count()
```
