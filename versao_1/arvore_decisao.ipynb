```python
#import os pacotes utilizados
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import re

#exibindo todas as colunas
pd.set_option('display.max_columns', None)

#definindo funcoes
def clean_string(string):
    cleanString = re.sub('\W+','', string)
    cleanString = re.sub(r'[^a-zA-Z0-9 ]',r'',cleanString)

    return cleanString
```

```python
dados = pd.read_json('dados/dados_flat_2.json')
print(dados.id.nunique())
print(dados.shape)
dados
```

```python
#criar novos campos de acordo com as disciplinas
for index1, group1 in dados.groupby(['descricao_historico_disciplina']):        

    dados[ clean_string(index1) ] = np.ones(dados.shape[0])*-1    
    #print(cleanString)
    
dados
```

```python
cont = 1

#loop sobre cada aluno preencher os atributos notas
for index1, group1 in dados.groupby(['id', 'descricao_historico_disciplina']):    
    #id do aluno
    id_aluno = index1[0]
    
    #nome da coluna relacionada a disciplina
    col_disc = clean_string(index1[1])    
    
    #nota relacionada a uma disciplina e um aluno
    nota = dados[(dados.id == id_aluno) & (dados.descricao_historico_disciplina == index1[1])]['media_final'].values 
    
    #para os casos do aluno ter ficado em recuperação só pegar apenas o primeiro registro
    #DEPOIS VERIFICAR OS CASOS QUE FICOU ASSIM
    nota = nota[0]
    
    #atribuindo nota a um aluno de uma disciplina
    dados.loc[dados.id == id_aluno, col_disc] = nota
    
    #print(dados.loc[dados.id == id_aluno, ['id', 'descricao_historico_disciplina', 'media_final', col_disc]])

    #if cont > 10:
    #    break
    #cont = cont + 1

print('drop duplicados')
    
X = dados.drop_duplicates('id')
X.shape
```
