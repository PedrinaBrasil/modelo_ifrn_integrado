# Merge entre os arquivos json

Realiza o merge dos dados (similiar ao join do sql), o qual, a partir de dois dataframes, é gerado um terceiro 
com os atributos dos dois anteriores ligados por algum campo em comum. 

Inicialmente a base de alunos possuem 11311 instâncias com 10 atributos, mas ao final é criado uma base de 188544 instâncias e  42 atributos. O aumento substancial da quantidade de instâncias se deve ao fato da repetição de alguns dados, por exemplo, um aluno que cursa 5 disciplinas será repetido 5 vezes na base, sendo cada linha representada por cada disciplina e com os dados do aluno em comum a todas. 

```python


#df_Alunos shape: (11311, 10)
print(df_Aluno.shape) 

#relacionando aluno ao seu campus
dados = pd.merge(df_Aluno, df_CursoCampus, 
                 left_on="curso_campus_id", right_on="id", suffixes=["", "_curso_campus"])
dados = pd.merge(dados, df_UnidadeOrganizacional, 
                 left_on="diretoria__setor__uo_id", right_on="id", suffixes=["", "_unidade_organizacional"])

#relacionando o aluno com sua situacao de matricula
dados = pd.merge(dados, df_SituacaoMatricula, 
                 left_on="situacao_id", right_on="id", suffixes=["", "_situacao_matricula"])
print(dados.shape)

#relacionando aluno aos dados de caracterizacao social
dados = pd.merge(dados, df_Caracterizacao, 
                 left_on="id", right_on="aluno_id", suffixes=["", "_caracterizacao"])
dados = pd.merge(dados, df_CompanhiaDomiciliar, 
                 left_on="companhia_domiciliar_id", right_on="id", suffixes=["", "_companhia_domiciliar"])
dados = pd.merge(dados, df_EstadoCivil, 
                 left_on="estado_civil_id", right_on="id", suffixes=["", "_estado_civil"])
dados = pd.merge(dados, df_TipoAcessoInternet, how='left', 
                 left_on="frequencia_acesso_internet_id", right_on="id", suffixes=["", "_internet"])
dados = pd.merge(dados, df_NivelEscolaridade, 
                 left_on="mae_nivel_escolaridade_id", right_on="id", suffixes=["", "_mae_escolaridade"])
dados = pd.merge(dados, df_NivelEscolaridade, 
                 left_on="pai_nivel_escolaridade_id", right_on="id", suffixes=["", "_pai_escolaridade"])
dados = pd.merge(dados, df_Raca, 
                 left_on="raca_id", right_on="id", suffixes=["", "_raca"])
dados = pd.merge(dados, df_RazaoAfastamentoEducacional, how='left', 
                 left_on="razao_ausencia_educacional_id", right_on="id", suffixes=["", "_afastamento"])
dados = pd.merge(dados, df_SituacaoTrabalho,
                 left_on="responsavel_financeir_trabalho_situacao_id", right_on="id", suffixes=["", "_trabalho_responsavel"])
dados = pd.merge(dados, df_ContribuinteRendaFamiliar, 
                 left_on="responsavel_financeiro_id", right_on="id", suffixes=["", "_responsavel_financeiro"])
dados = pd.merge(dados, df_NivelEscolaridade, 
                 left_on="responsavel_financeiro_nivel_escolaridade_id", right_on="id", suffixes=["", "_responsavel_escolaridade"])
dados = pd.merge(dados, df_TipoAreaResidencial, 
                 left_on="tipo_area_residencial_id", right_on="id", suffixes=["", "_area_residencial"])
dados = pd.merge(dados, df_TipoImovelResidencial, 
                 left_on="tipo_imovel_residencial_id", right_on="id", suffixes=["", "_imovel"])
dados = pd.merge(dados, df_TipoServicoSaude, how='left', 
                 left_on="tipo_servico_saude_id", right_on="id", suffixes=["", "_saude"])
dados = pd.merge(dados, df_SituacaoTrabalho,  
                 left_on="trabalho_situacao_id", right_on="id", suffixes=["", "_trabalho"])
print(dados.shape)
dados.head(2)


#relacionando aluno a situacao da matricula naquele periodo
dados = pd.merge(dados, df_MatriculaPeriodo, 
                 left_on="id", right_on="aluno_id", suffixes=["", "_matricula_periodo"])
dados = pd.merge(dados, df_SituacaoMatriculaPeriodo, 
                 left_on="situacao_id_matricula_periodo", right_on="id", suffixes=["", "_situacao_matricula_periodo"])
print(dados.shape)
dados.head(2)

#relacionando aluno a nota das disciplinas daquele periodo
dados = pd.merge(dados, df_Notas, 
                 left_on="id_matricula_periodo", right_on="matricula_periodo_id", suffixes=["", "_notas"])
dados = pd.merge(dados, df_Disciplina, 
                 left_on="disciplina_id", right_on="id", suffixes=["", "_disciplina"])

print(dados.shape)
#print(dados.head(2))

#selecionando os campos no dataframe
dados = dados[['id', 'ano_letivo__ano', 'cep', 'pessoa_fisica__nascimento_data', 
               #'cidade__nome', 'cidade__estado__nome', #INFORMACAO REDUNDANTE EM RELACAO AO CEP
               'pessoa_fisica__sexo', 'descricao_historico', 'sigla',
               'descricao', 
               'ano_letivo__ano_matricula_periodo', 'descricao_situacao_matricula_periodo',
               #'data', 'descricao_encaminhamento', #FUTURAMENTE SERA UTILIZADO
               'media_final', 'percentual_frequencia', 
               'descricao_historico_disciplina',
               #'professores', #FUTURAMENTE SERA UTILIZADO
               'aluno_exclusivo_rede_publica', 
               'descricao_companhia_domiciliar', 'descricao_estado_civil',
               'ficou_tempo_sem_estudar', 'descricao_internet', 
               'local_acesso_internet', 'descricao_mae_escolaridade', 
               'nome_escola_ensino_fundamental', 'nome_escola_ensino_medio', 
               'descricao_pai_escolaridade', 'possui_conhecimento_informatica',
               'possui_necessidade_especial', 'qtd_filhos', 'qtd_pessoas_domicilio', 
               'quantidade_computadores', 'quantidade_netbooks', 'quantidade_notebooks',
               'quantidade_smartphones', 'descricao_raca', 'descricao_afastamento', 
               'renda_bruta_familiar', 'descricao_trabalho_responsavel', 'descricao_responsavel_financeiro', 
               'descricao_responsavel_escolaridade', 'tempo_sem_estudar',
               'descricao_area_residencial', 'descricao_imovel', 'descricao_saude',               
               'descricao_trabalho'
              ]]


"""
#SEM CARACTERIZACAO SOCIAL
dados = dados[['id', 'ano_letivo__ano', 'cep', 'pessoa_fisica__nascimento_data', 
               #'cidade__nome', 'cidade__estado__nome', 
               'pessoa_fisica__sexo', 'descricao_historico', 'sigla',
               'descricao', 
               'ano_letivo__ano_matricula_periodo', 'descricao_situacao_matricula_periodo',
               #'data', 'descricao_encaminhamento',
               'media_final', 'percentual_frequencia', 
               'descricao_historico_disciplina',
               #'professores', #FUTURAMENTE SERA UTILIZADO
               #'aluno_exclusivo_rede_publica', 
               #'descricao_companhia_domiciliar', 'descricao_estado_civil',
               #'ficou_tempo_sem_estudar', 'descricao_internet', 
               #'local_acesso_internet', 'descricao_mae_escolaridade', 
               #'nome_escola_ensino_fundamental', 'nome_escola_ensino_medio', 
               #'descricao_pai_escolaridade', 'possui_conhecimento_informatica',
               #'possui_necessidade_especial', 'qtd_filhos', 'qtd_pessoas_domicilio', 
               #'quantidade_computadores', 'quantidade_netbooks', 'quantidade_notebooks',
               #'quantidade_smartphones', 'descricao_raca', 'descricao_afastamento', 
               #'renda_bruta_familiar', 'descricao_trabalho_responsavel', 'descricao_responsavel_financeiro', 
               #'descricao_responsavel_escolaridade', 'tempo_sem_estudar',
               #'descricao_area_residencial', 'descricao_imovel', 'descricao_saude',               
               #'descricao_trabalho'
              ]]
"""

print(dados.shape)
dados.head(2)
```

# Transformando as notas em colunas

a fim de simplificar a base e remover dados duplicados, foram criados 4 colunas referentes as notas de português e matemática do 1º e 2º ano

```python
print(dados.id.nunique())
print(dados.shape)
```

```python
#definindo funcoes
import re
def clean_string(string):
    cleanString = re.sub('\W+','', string)
    cleanString = re.sub(r'[^a-zA-Z0-9 ]',r'',cleanString)

    return cleanString
```

```python
#criar novos campos de acordo com as disciplinas
for index1, group1 in dados.groupby(['descricao_historico_disciplina']):        

    dados[ clean_string(index1) ] = np.ones(dados.shape[0])*-1    
    #print(cleanString)
    
dados.head(2)
```

```python
cont = 1

#loop sobre cada aluno preencher os atributos notas
for index1, group1 in dados.groupby(['id', 'descricao_historico_disciplina']):    
    #id do aluno
    id_aluno = index1[0]
    
    #nome da coluna relacionada a disciplina
    col_disc = clean_string(index1[1])    
    
    #nota relacionada a uma disciplina e um aluno
    nota = dados[(dados.id == id_aluno) & (dados.descricao_historico_disciplina == index1[1])]['media_final'].values 
    
    #para os casos do aluno ter ficado em recuperação só pegar apenas o primeiro registro
    #DEPOIS VERIFICAR OS CASOS QUE FICOU ASSIM
    nota = nota[0]
    
    #atribuindo nota a um aluno de uma disciplina
    dados.loc[dados.id == id_aluno, col_disc] = nota
    
    #print(dados.loc[dados.id == id_aluno, ['id', 'descricao_historico_disciplina', 'media_final', col_disc]])

    #if cont > 10:
    #    break
    #cont = cont + 1


    
#removendo duplicadas
X = dados.drop_duplicates('id')

#removendo atributos já expressos em outras colunas
X = X.drop(['media_final', 'descricao_historico_disciplina'], axis=1)

print(X.shape)
X.head(2)
```

```python
#removendo alunos que não possuam todas as notas do primeiro ano
df = X[X.LnguaPortuguesaeLiteraturaI90H != -1]
print(df.shape)

df = df[df.MatemticaI120H != -1]
print(df.shape)

#df = df[df.LnguaPortuguesaeLiteraturaII90H != -1]
#print(df.shape)

#df = df[df.MatemticaII90H != -1]
#print(df.shape)

dados.head(1)

#dados = df
```
