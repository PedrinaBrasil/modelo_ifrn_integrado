```python
#definindo funcoes
import re
def clean_string(string):
    cleanString = re.sub('\W+','', string)
    cleanString = re.sub(r'[^a-zA-Z0-9 ]',r'',cleanString)

    return cleanString

from datetime import datetime
from datetime import date, timedelta

def get_age(born):
    b = datetime.strptime(born, "%d/%m/%Y")
    age = (date.today() - b.date()) // timedelta(days=365.2425)
    return age

def get_age_df(df):
    b = datetime.strptime(df['pessoa_fisica__nascimento_data'], "%d/%m/%Y")
    ingresso = datetime.strptime(str(df['ano_letivo__ano']), "%Y")
    age = (ingresso.date() - b.date()) // timedelta(days=365.2425)
    
    return age
```

# Merge entre os arquivos json

Realiza o merge dos dados (similiar ao join do sql), o qual, a partir de dois dataframes, é gerado um terceiro 
com os atributos dos dois anteriores ligados por algum campo em comum. 

```python
#df_Alunos shape: (11311, 10)
print(df_Aluno.shape) 

#relacionando aluno ao seu campus
dados = pd.merge(df_Aluno, df_CursoCampus, 
                 left_on="curso_campus_id", right_on="id", suffixes=["", "_curso_campus"])
dados = pd.merge(dados, df_UnidadeOrganizacional, 
                 left_on="diretoria__setor__uo_id", right_on="id", suffixes=["", "_unidade_organizacional"])

#relacionando o aluno com sua situacao de matricula
dados = pd.merge(dados, df_SituacaoMatricula, 
                 left_on="situacao_id", right_on="id", suffixes=["", "_situacao_matricula"])
print(dados.shape)

#relacionando aluno aos dados de caracterizacao social
dados = pd.merge(dados, df_Caracterizacao, 
                 left_on="id", right_on="aluno_id", suffixes=["", "_caracterizacao"])
dados = pd.merge(dados, df_CompanhiaDomiciliar, 
                 left_on="companhia_domiciliar_id", right_on="id", suffixes=["", "_companhia_domiciliar"])
dados = pd.merge(dados, df_EstadoCivil, 
                 left_on="estado_civil_id", right_on="id", suffixes=["", "_estado_civil"])
dados = pd.merge(dados, df_TipoAcessoInternet, how='left', 
                 left_on="frequencia_acesso_internet_id", right_on="id", suffixes=["", "_internet"])
dados = pd.merge(dados, df_NivelEscolaridade, 
                 left_on="mae_nivel_escolaridade_id", right_on="id", suffixes=["", "_mae_escolaridade"])
dados = pd.merge(dados, df_NivelEscolaridade, 
                 left_on="pai_nivel_escolaridade_id", right_on="id", suffixes=["", "_pai_escolaridade"])
dados = pd.merge(dados, df_Raca, 
                 left_on="raca_id", right_on="id", suffixes=["", "_raca"])
dados = pd.merge(dados, df_RazaoAfastamentoEducacional, how='left', 
                 left_on="razao_ausencia_educacional_id", right_on="id", suffixes=["", "_afastamento"])
dados = pd.merge(dados, df_SituacaoTrabalho,
                 left_on="responsavel_financeir_trabalho_situacao_id", right_on="id", suffixes=["", "_trabalho_responsavel"])
dados = pd.merge(dados, df_ContribuinteRendaFamiliar, 
                 left_on="responsavel_financeiro_id", right_on="id", suffixes=["", "_responsavel_financeiro"])
dados = pd.merge(dados, df_NivelEscolaridade, 
                 left_on="responsavel_financeiro_nivel_escolaridade_id", right_on="id", suffixes=["", "_responsavel_escolaridade"])
dados = pd.merge(dados, df_TipoAreaResidencial, 
                 left_on="tipo_area_residencial_id", right_on="id", suffixes=["", "_area_residencial"])
dados = pd.merge(dados, df_TipoImovelResidencial, 
                 left_on="tipo_imovel_residencial_id", right_on="id", suffixes=["", "_imovel"])
dados = pd.merge(dados, df_TipoServicoSaude, how='left', 
                 left_on="tipo_servico_saude_id", right_on="id", suffixes=["", "_saude"])
dados = pd.merge(dados, df_SituacaoTrabalho,  
                 left_on="trabalho_situacao_id", right_on="id", suffixes=["", "_trabalho"])
print(dados.shape)
#dados.head(2)

#selecionando os campos no dataframe
dados = dados[['id', 
               'ano_letivo__ano', 
               'cep', 
               'pessoa_fisica__nascimento_data', 
               'pessoa_fisica__sexo', 
               'descricao_historico', 
               'sigla',
               'descricao', 
               'aluno_exclusivo_rede_publica',
               'ensino_fundamental_conclusao', 
               'ficou_tempo_sem_estudar',
               'local_acesso_internet', 
               'possui_conhecimento_informatica', 
               'possui_necessidade_especial', 
               'qtd_pessoas_domicilio', 
               'quantidade_computadores', 
               'quantidade_netbooks', 
               'quantidade_notebooks', 
               'quantidade_smartphones', 
               'renda_bruta_familiar',
               'descricao_companhia_domiciliar', 
               'descricao_estado_civil', 
               'descricao_internet', 
               'descricao_mae_escolaridade', 
               'descricao_pai_escolaridade', 
               'descricao_raca', 
               'descricao_responsavel_financeiro', 
               'descricao_responsavel_escolaridade', 
               'descricao_area_residencial', 
               'descricao_imovel', 
               'descricao_saude', 
               'descricao_trabalho', 
              ]]

print(dados.id.nunique())
print(dados.shape)
dados.head(3)
```

# Criando atributo classe

```python
#agrupando os alunos cancelados, evadidos, transferidos interno e externo como evasão para o mec
#-1 valor sentinela, 0 evasão, 1 matriculado
dados['classe'] = np.ones(dados.shape[0])*-1

dados.loc[dados['descricao'].isin(['Evasão',
                                   'Cancelado', 
                                   'Cancelamento Compulsório', 
                                   #'Transferido Externo', #PARA O MEC É CONSIDERADO EVASAO MAS NAO SERA CONSIDERADO NO MODELO
                                   #'Transferido Interno'  #PARA O MEC É CONSIDERADO EVASAO MAS NAO SERA CONSIDERADO NO MODELO
                                  ]), 'classe'] = 0

dados.loc[dados['descricao'].isin(['Concluído',
                                   #'Intercâmbio', #PARA O MEC NÃO É EVASAO MAS NAO SERA CONSIDERADO NO MODELO
                                   'Matriculado', 
                                   #'Trancado', #PARA O MEC NÃO É EVASAO MAS NAO SERA CONSIDERADO NO MODELO
                                   #'Trancado Voluntariamente', #PARA O MEC NÃO É EVASAO MAS NAO SERA CONSIDERADO NO MODELO
                                   'Matrícula Vínculo Institucional']), 'classe'] = 1
```

# Criando atributo notas

```python
#relacionando aluno a situacao da matricula naquele periodo
dados_notas = pd.merge(dados, df_MatriculaPeriodo, 
                 left_on="id", right_on="aluno_id", suffixes=["", "_matricula_periodo"])

dados_notas = pd.merge(dados_notas, df_SituacaoMatriculaPeriodo, 
                 left_on="situacao_id", right_on="id", suffixes=["", "_situacao_matricula_periodo"])
print(dados_notas.shape)

#relacionando aluno a nota das disciplinas daquele periodo
dados_notas = pd.merge(dados_notas, df_Notas, 
                 left_on="id_matricula_periodo", right_on="matricula_periodo_id", suffixes=["", "_notas"])
dados_notas = pd.merge(dados_notas, df_Disciplina, 
                 left_on="disciplina_id", right_on="id", suffixes=["", "_disciplina"])

dados_notas = dados_notas[['id', 'descricao_historico_disciplina', 'percentual_frequencia', 'media_final']]

print(dados_notas.shape)
dados_notas.head(2)
```

serão apenas utilizados as notas de português e matemática do 1º ano

```python
dados_filtrado = dados_notas[(dados_notas.descricao_historico_disciplina.str.startswith('Língua Portuguesa')) | 
              (dados_notas.descricao_historico_disciplina.str.startswith('Matemática I')) ]
print(dados_filtrado.shape)
dados_filtrado.head(2)


dados_filtrado["descricao_historico_disciplina"].value_counts()
```

```python
dados[ clean_string('Matemática I(120H)') ] = np.ones(dados.shape[0])*-1 
dados[ clean_string('Matemática I(120H)') + '_freq' ] = np.ones(dados.shape[0])*-1
dados[ clean_string('Matemática I(120H)') + '_dependencia' ] = np.ones(dados.shape[0])*-1

dados[ clean_string('Língua Portuguesa e Literatura I(90H)') ] = np.ones(dados.shape[0])*-1 
dados[ clean_string('Língua Portuguesa e Literatura I(90H)') + '_freq' ] = np.ones(dados.shape[0])*-1 
dados[ clean_string('Língua Portuguesa e Literatura I(90H)') + '_dependencia' ] = np.ones(dados.shape[0])*-1 

dados[ 'idade' ] = np.ones(dados.shape[0])*-1 
```

```python
#para cada aluno recupera a nota de Português e Matemática do 1º ano. Caso o mesmo tenha tido dependência salva a nota mais alta
for i in dados.id:
    
    #recupera a nota de matematica e atribui ao aluno
    r_mat = dados_filtrado[(dados_filtrado.id == i) & (dados_filtrado.descricao_historico_disciplina == 'Matemática I(120H)')]
    dados.loc[dados.id == i, clean_string('Matemática I(120H)') ] = r_mat.media_final.max()
    dados.loc[dados.id == i, clean_string('Matemática I(120H)') + '_freq' ] = r_mat.percentual_frequencia.max()  
    dados.loc[dados.id == i, clean_string('Matemática I(120H)') + '_dependencia' ] = r_mat.shape[0] - 1

    #recupera a nota de portugues e atribui ao aluno
    r_por = dados_filtrado[(dados_filtrado.id == i) & 
                           (dados_filtrado.descricao_historico_disciplina == 'Língua Portuguesa e Literatura I(90H)')]
    dados.loc[dados.id == i, clean_string('Língua Portuguesa e Literatura I(90H)') ] = r_por.media_final.max() 
    dados.loc[dados.id == i, clean_string('Língua Portuguesa e Literatura I(90H)') + '_freq' ] = r_por.percentual_frequencia.max() 
    dados.loc[dados.id == i, clean_string('Língua Portuguesa e Literatura I(90H)') + '_dependencia' ] = r_mat.shape[0] - 1 
    
dados.head(3)
    
```

```python
#verificando a média de idade dos alunos que entram no ano_letivo de 2015, 2016 e 2017.
#o resultado mostra que a média dos alunos que entram nesse período é de 15
print(dados[dados.ano_letivo__ano == 2015].idade.mean())
print(dados[dados.ano_letivo__ano == 2016].idade.mean())
print(dados[dados.ano_letivo__ano == 2017].idade.mean())

df_tmp = dados.copy()

#atribui NaN para os dados que nao tem registrado o ano de ensino_fundamental
df_tmp['tempo_entre_conclusao_ingresso'] = df_tmp['tempo_entre_conclusao_ingresso'].replace([2015, 2016, 2017], np.NaN)

#verifica media e desvio padrao
print('Média e Std para 15 anos')
print(df_tmp[(df_tmp.idade == 15) & (df_tmp.idade == 15)].tempo_entre_conclusao_ingresso.mean())
print(df_tmp[(df_tmp.idade == 15) & (df_tmp.idade == 15)].tempo_entre_conclusao_ingresso.std())

#verifica a quantidade para cada tempo
print('Qnt de tempo para cada valor com idade de 15 anos')
print(df_tmp[(df_tmp.idade == 15) & (df_tmp.idade == 15)].tempo_entre_conclusao_ingresso.value_counts())

#media do tempo entre conclusao e ingressos dos alunos com 15 anos
#media_tempo_15 = round(df_tmp[df_tmp.idade == 15].tempo_entre_conclusao_ingresso.mean())

#preenche com -1 os NaN
df_tmp['tempo_entre_conclusao_ingresso'].fillna(-1, inplace=True)
print(df_tmp['tempo_entre_conclusao_ingresso'].value_counts())

#remove os que forem tempo igual a media * 2 sigma
print("Limpando")
print(df_tmp.shape)
df_tmp = df_tmp[df_tmp.tempo_entre_conclusao_ingresso < (df_tmp['tempo_entre_conclusao_ingresso'].median() + 2*df_tmp['tempo_entre_conclusao_ingresso'].std())]
print(df_tmp.shape)

#verifica as probabilidades de cada tempo
print("gerando prob")
prob = (df_tmp[(df_tmp.tempo_entre_conclusao_ingresso >= 0) & 
               (df_tmp.idade == 15)].tempo_entre_conclusao_ingresso.value_counts()) / sum((df_tmp[(df_tmp.idade == 15) & 
                                                            (df_tmp.tempo_entre_conclusao_ingresso >= 0)].tempo_entre_conclusao_ingresso.value_counts()))
print(prob)
print(sum(prob))

#atualiza os valores -1 para a media calculada de acordo com os alunos que tem 15 anos
#df_tmp['tempo_entre_conclusao_ingresso'] = df_tmp['tempo_entre_conclusao_ingresso'].replace([-1], media_tempo_15)

print('antes')
print(df_tmp['tempo_entre_conclusao_ingresso'].value_counts())

#tamanho da amostra com -1
tam = df_tmp[df_tmp.tempo_entre_conclusao_ingresso == -1].tempo_entre_conclusao_ingresso.shape[0]

#gera lista de acordo com as probabilidades de cada tempo
res = np.random.choice( [0, 1, 2, 3], tam, p=[ prob[0], prob[1], prob[2], prob[3] ] )

#definie que essas linhas foram feitas de forma artificial
df_tmp.loc[df_tmp.tempo_entre_conclusao_ingresso == -1, 'artificial'] = 1

#atribui os tempos de acordo com as probabilidades geradas se baseando na idade de 15 anos
df_tmp.loc[df_tmp.tempo_entre_conclusao_ingresso == -1, 'tempo_entre_conclusao_ingresso'] = res

print(df_tmp['tempo_entre_conclusao_ingresso'].value_counts())
print(df_tmp['artificial'].value_counts())

#atualizando dados 
dados = df_tmp
```

```python
#import os pacotes utilizados
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

#selecionando atributos 
from sklearn.feature_selection import RFECV

#ESCALONAMENTO DE DADOS
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler

#SELECIONANDO MELHOR MODELO
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from scipy.stats import randint as sp_randint
from sklearn.cross_validation import train_test_split
from sklearn.model_selection import cross_val_score

#Evaluating
import seaborn as sns; sns.set()

from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score 
from sklearn.metrics import precision_score, recall_score 
from sklearn.metrics import f1_score

#EXPORTAR ARVORE
from sklearn.tree import export_graphviz
import os

#CURVA ROC
%matplotlib inline
from sklearn.metrics import roc_curve
from sklearn.model_selection import cross_val_predict 
from sklearn.metrics import roc_auc_score

#REDUÇÃO DE DIMENSIONALIDADE
from sklearn.decomposition import PCA

#exibindo todas as colunas
pd.set_option('display.max_columns', None)

#definicao de funcoes que serao utilizadas
# Utility function to report best scores
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
            
def plot_roc_curve(fpr, tpr, label=None):    
    plt.plot(fpr, tpr, linewidth=2, label=label)    
    plt.plot([0, 1], [0, 1], 'k--')    
    plt.axis([0, 1, 0, 1])   
    plt.xlabel('False Positive Rate')    
    plt.ylabel('True Positive Rate')
    
#SELECIONANDO MELHOR MODEL RF
def make_RF(Xtrain, ytrain):
    #cria a randomforest 
    #forest_clf = RandomForestClassifier(oob_score=True, n_jobs=-1) 
    forest_clf = RandomForestClassifier(n_jobs=-1) 

    param_grid_forest = {'max_depth': [3, 5, None],
                  'max_features': ['auto', 'sqrt', 'log2'],
                  'n_estimators': [10, 30, 60, 100, 200],
                  'criterion': ["gini", "entropy"]}

    #seleciona o melhor modelo de acordo com os parâmetros
    grid_search_forest = GridSearchCV(estimator=forest_clf, param_grid=param_grid_forest)
    grid_search_forest.fit(Xtrain, ytrain)
    print(grid_search_forest.best_params_)
    print(grid_search_forest.best_estimator_)
    
    #mostra os 5 melhores modelos
    report(grid_search_forest.cv_results_, 5)    
    
    return grid_search_forest.best_estimator_

def evaluating(model, ytest, ypred, X, y, Xtest):
    #Evaluating
    mat = confusion_matrix(ytest, ypred)
    sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
    plt.xlabel('true label')
    plt.ylabel('predicted label');

    #evaluating
    print(precision_score(ytest, ypred))
    print(recall_score(ytest, ypred))

    #evaluating
    print(precision_score(ytest, ypred))
    print(recall_score(ytest, ypred))
    print(f1_score(ytest, ypred))

    #qnt de evadidos 
    print(sum(ytest==0))

    #precisao
    print('Precisao')
    scores = cross_val_score(model, X, y, cv=10)
    accuracy = np.mean(scores)
    print(scores)
    print(accuracy)

    #CURVA ROC
    fig = plt.figure()
    y_probas = cross_val_predict(model, Xtest, ytest, cv=10, method="predict_proba") 

    y_scores = y_probas[:, 1]   # score = proba of positive class 
    fpr, tpr, thresholds = roc_curve(ytest, y_scores)

    plot_roc_curve(fpr, tpr) 
    plt.show()
    print(roc_auc_score(ytest, y_scores))

    #exibe os atributos mais importantes
    importances = model.feature_importances_
    std = np.std([tree.feature_importances_ for tree in model.estimators_], axis=0)
    indices = np.argsort(importances)[::-1]

    # Print the feature ranking
    print("Feature ranking:")

    for f in range(Xtest.shape[1]):
        print("%d. feature %d [%s] (%f)" % (f + 1, indices[f], X.columns[indices[f]],importances[indices[f]]))

    # Plot the feature importances of the forest
    plt.figure()
    plt.title("Feature importances")
    plt.bar(range(Xtest.shape[1]), importances[indices], color="r", yerr=std[indices], align="center")
    plt.xticks(range(Xtest.shape[1]), indices)
    plt.xlim([-1, Xtest.shape[1]])
    plt.show()    

#limpando os NaN devido ao RF
def preparing_data(dados):
    print(dados.shape)
    dados = dados.dropna()
    print(dados.shape)
    
    #configura os dados de entrada e saída
    y = dados.classe
    X = dados

    del X['classe']
    X.shape    
    
    #binazira os dados devido aos atributos categoricos 
    dados_dummies = pd.get_dummies(X, drop_first=True)
    dados_dummies.shape

    #remove atributos inconistentes
    dados_clean = dados_dummies.replace([np.inf, -np.inf], np.nan)
    dados_clean = dados_clean.dropna(axis=1, how='all')
    print(dados_clean.info())
    
    return (dados_clean, y)
```
